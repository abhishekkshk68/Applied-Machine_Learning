```python
import os
CATEGORIES = ["dog", "cat"]
DataDir="/content/"
features=[]
label=[]
for category in CATEGORIES:  # do dogs and cats
    path = os.path.join(DataDir,category) 
    print(path) # create path to dogs and cats
    class_num = CATEGORIES.index(category)
    for img in os.listdir(path):  # iterate over each image per dogs and cats
        print(img)
        img12=Image.open(DataDir+category+"/"+img)
        image_file = img12.convert('1').resize((28, 28))
        data= asarray(image_file).reshape(1,-1)
        features.append(list(data))
        label.append(class_num)
        # convert to array
        #print(image_file)
#label.reshape(-1,1)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(features, label, test_size=0.50, random_state=1)
```

```python
model=GaussianNB()
print(X_train)
print(y_train)
```

```python
model.fit(X_train,X_test)
```
